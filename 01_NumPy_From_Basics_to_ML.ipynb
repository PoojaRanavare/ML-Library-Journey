# 1. Introduction to NumPy

## What is NumPy?

Imagine you have a notebook where you write thousands or millions of numbers.

A normal calculator will take a lot of time to calculate all these numbers.

NumPy solves this problem.

## NumPy (Numerical Python)

NumPy is a Python library used for:

✅ Fast mathematical calculations  
✅ Handling large numerical data  
✅ Creating arrays and matrices  
✅ Supporting Machine Learning algorithms

### Simple Definition:

> NumPy is a super-fast calculator for Python that can work with millions of numbers together.

# 3.Why NumPy was Created?

Before NumPy, programmers used Python lists.

Example:

numbers = [1,2,3,4,5]

But when data became huge:

- Images
- Scientific calculations
- Machine Learning datasets
- Millions of numbers

Python became slower.

NumPy provides:

✅ Faster processing  
✅ Less memory usage  
✅ Easy mathematical operations  
✅ Matrix handling

#11. Array Functions

#1. Array Creation Functions
| Function        | Purpose                        | Example               |
| --------------- | ------------------------------ | --------------------- |
| `np.array()`    | Create NumPy array from list   | `np.array([1,2,3])`   |
| `np.zeros()`    | Create array filled with zeros | `np.zeros(5)`         |
| `np.ones()`     | Create array filled with ones  | `np.ones(5)`          |
| `np.empty()`    | Create empty array             | `np.empty(3)`         |
| `np.arange()`   | Create sequence of numbers     | `np.arange(1,10)`     |
| `np.linspace()` | Create evenly spaced numbers   | `np.linspace(0,10,5)` |
| `np.full()`     | Create array with same value   | `np.full((2,2),7)`    |


```python

import numpy as np

arr = np.zeros(5)

print(arr)
```

#2. Array Information Functions
| Function | Purpose              |
| -------- | -------------------- |
| `ndim`   | Number of dimensions |
| `shape`  | Size of array        |
| `size`   | Total elements       |
| `dtype`  | Data type            |


```python
arr=np.array([[1,2,3],[4,5,6]])

print(arr.ndim)
print(arr.shape)
print(arr.size)
print(arr.dtype)
```

#3. Mathematical Functions
| Function      | Purpose            |
| ------------- | ------------------ |
| `np.sum()`    | Addition of values |
| `np.mean()`   | Average            |
| `np.median()` | Middle value       |
| `np.min()`    | Minimum value      |
| `np.max()`    | Maximum value      |
| `np.std()`    | Standard deviation |
| `np.var()`    | Variance           |


```python
marks=np.array([80,90,70,85])

print(np.mean(marks))
print(np.max(marks))
```

#4. Array Manipulation Functions
| Function      | Purpose                         |
| ------------- | ------------------------------- |
| `reshape()`   | Change array shape              |
| `flatten()`   | Convert multi-dimensional to 1D |
| `transpose()` | Change rows to columns          |
| `resize()`    | Change size                     |


```python
arr=np.array([1,2,3,4])

new_arr=arr.reshape(2,2)

print(new_arr)
```
