# What is TensorFlow?

TensorFlow is an open-source Python library developed for Artificial Intelligence (AI) and Deep Learning.

It helps computers learn from data and make predictions without being explicitly programmed for every task.

Using TensorFlow, we can:

✔ Build Deep Learning Models

✔ Train Neural Networks

✔ Image Recognition

✔ Face Recognition

✔ Speech Recognition

✔ Language Translation

✔ Recommendation Systems

✔ Time Series Prediction

TensorFlow is one of the world's most popular Deep Learning libraries.

# History of TensorFlow

📅 First Released:
2015

🏢 Developed By:
Google Brain Team

TensorFlow was developed by the Google Brain team to make Artificial Intelligence and Machine Learning research easier.

It replaced Google's earlier deep learning system called DistBelief and was later released as open-source software.

Today, TensorFlow is widely used in AI research, industries, and universities.

# Why was TensorFlow Created?

Traditional Machine Learning works well for many tasks, but complex problems such as image recognition, speech recognition, and natural language processing require Deep Learning.

TensorFlow was created to:

✔ Build Neural Networks

✔ Train Deep Learning Models

✔ Handle Large Datasets

✔ Run on CPUs, GPUs, and TPUs

✔ Develop AI Applications Faster

# Create a Constant Tensor

```python
import tensorflow as tf

tensor = tf.constant([10,20,30])           # creates immutable tensor

print(tensor)
```

# Create a Variable

```python
var = tf.Variable([5,10,15])

print(var)
```

# Tensor Shape

```python
print(tensor.shape)
```

# Tensor Data Type

```python
print(tensor.dtype)
```

# Tensor Addition

```python
a = tf.constant([1,2,3])

b = tf.constant([4,5,6])

print(tf.add(a,b))
```

# Tensor Multiplication

```python
print(tf.multiply(a,b))
```

# Matrix Multiplication

```python
A = tf.constant([[1,2],[3,4]])

B = tf.constant([[5,6],[7,8]])

print(tf.matmul(A,B))
```

# Random Tensor

```python
random_tensor = tf.random.uniform((2,3))

print(random_tensor)
```

# Reshape Tensor

```python
tensor = tf.constant([1,2,3,4,5,6])

print(tf.reshape(tensor,(2,3)))
```

# Convert NumPy to Tensor

```python
import numpy as np

arr = np.array([1,2,3])

tensor = tf.convert_to_tensor(arr)

print(tensor)
```

# Convert Tensor to NumPy

```python
print(tensor.numpy())
```
