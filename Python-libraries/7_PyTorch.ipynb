# What is PyTorch?

PyTorch is an open-source Python library used for Artificial Intelligence (AI), Machine Learning (ML), and Deep Learning (DL).

It helps developers build and train neural networks for solving complex problems.

Using PyTorch, we can:

✔ Build Deep Learning Models

✔ Image Classification

✔ Object Detection

✔ Natural Language Processing (NLP)

✔ Speech Recognition

✔ Chatbots

✔ Recommendation Systems

✔ Robotics

PyTorch is widely used in research and industry because it is flexible and easy to debug.

# History of PyTorch

📅 First Released:
2016

🏢 Developed By:
Facebook AI Research (FAIR)

👨 Main Contributors:
Adam Paszke and the Facebook AI Research team

PyTorch was developed to make Deep Learning easier and more flexible.

Today, it is maintained by the PyTorch Foundation under the Linux Foundation and is used by researchers, universities, and technology companies around the world.

# Why was PyTorch Created?

Earlier Deep Learning frameworks were powerful but often difficult to debug and experiment with.

PyTorch was created to provide:

✔ Easy model building

✔ Dynamic computation graphs

✔ Faster research and experimentation

✔ GPU acceleration

✔ Better debugging using normal Python code

# Features of PyTorch

# Import PyTorch

```python
# Import PyTorch

import torch          # torch is the main package of PyTorch.
```

# Create a Tensor

```python
import torch

tensor = torch.tensor([10,20,30])        # Creates a Pytorch tensor

print(tensor)
```

# Tensor Shape

```python
print(tensor.shape)
```

# Tensor Data Type

```python
print(tensor.dtype)
```

# Create Zeros Tensor

```python
tensor = torch.zeros((2,3))

print(tensor)
```

# Create Ones Tensor

```python
tensor = torch.ones((2,3))

print(tensor)
```

# Random Tensor

```python
tensor = torch.rand((2,3))

print(tensor)
```

# Tensor Addition

```python
a = torch.tensor([1,2,3])

b = torch.tensor([4,5,6])

print(a+b)
```

# Tensor Multiplication

```python
print(a*b)
```

# Matrix Multiplication

```python
A = torch.tensor([[1,2],[3,4]])

B = torch.tensor([[5,6],[7,8]])

print(torch.matmul(A,B))
```

# Reshape Tensor

```python
tensor = torch.arange(6)

print(tensor.reshape(2,3))
```

# Convert NumPy Array to Tensor

```python
import numpy as np

arr = np.array([1,2,3])

tensor = torch.from_numpy(arr)

print(tensor)
```

# Convert Tensor to NumPy

```python
print(tensor.numpy())
```

# Check GPU Availability

```python
print(torch.cuda.is_available())  # Returns True if a CUDA-compatible GPU is available.
```

# Move Tensor to GPU

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

tensor = tensor.to(device)

print(tensor)
```

# Create a Neural Network

```python
import torch.nn as nn

model = nn.Sequential(
    nn.Linear(4,16),
    nn.ReLU(),
    nn.Linear(16,3)
)

print(model)
```

# Loss Function

```python
criterion = nn.CrossEntropyLoss()
```

# Optimizer

```python
import torch.optim as optim

optimizer = optim.Adam(model.parameters(), lr=0.001)
```

# Forward Pass

```python
x = torch.rand((1,4))

output = model(x)

print(output)
```
