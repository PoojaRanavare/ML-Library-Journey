# What is Keras?

Keras is an open-source Python library used for building Deep Learning and Artificial Intelligence models.

It provides a simple and user-friendly interface to create Neural Networks without writing complex code.

Keras helps us:

✔ Build Neural Networks

✔ Train Deep Learning Models

✔ Classify Images

✔ Predict Future Values

✔ Process Text

✔ Recognize Speech

✔ Detect Objects

Today, Keras is included inside TensorFlow and is used through **tensorflow.keras**.

# History of Keras

📅 First Released:
2015

👨 Creator:
François Chollet

🏢 Developed By:
Google

Keras was created by François Chollet to make Deep Learning easier and more accessible.

Since TensorFlow 2.0 (2019), Keras has become TensorFlow's official high-level API.

Today, millions of developers use Keras for AI and Deep Learning projects.

# Why was Keras Created?

Earlier Deep Learning libraries required writing a lot of complex code.

Keras was created to:

✔ Make Deep Learning simple

✔ Reduce coding effort

✔ Help beginners learn AI easily

✔ Build models quickly

✔ Improve readability and productivity

# Import Keras

# Create a Sequential Model

```python
from tensorflow.keras.models import Sequential

model = Sequential()                    # Creates an empty neural network model.

print(model)
```

# Add Layers

```python
from tensorflow.keras.layers import Dense

model.add(Dense(16, activation="relu"))

model.add(Dense(8, activation="relu"))

model.add(Dense(1))
```

# Build Model in One Step

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([
    Dense(16, activation="relu"),
    Dense(8, activation="relu"),
    Dense(1)
])
```

# View Model Summary

```python
model.summary()
```

# Compile Model

```python
model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["accuracy"]
)
```

# Dropout Layer

```python
from tensorflow.keras.layers import Dropout

model.add(Dropout(0.2))         # Helps reduce overfitting by randomly disabling some neurons during training.
```

# Flatten Layer

```python
from tensorflow.keras.layers import Flatten

model.add(Flatten())
```

# Conv2D Layer

```python
from tensorflow.keras.layers import Conv2D

model.add(
    Conv2D(
        filters=32,
        kernel_size=(3,3),
        activation="relu"
    )
)
```

# MaxPooling Layer

```python
from tensorflow.keras.layers import MaxPooling2D

model.add(MaxPooling2D(pool_size=(2,2)))
```
