# What is Seaborn?

Seaborn is an open-source Python library used for creating beautiful and informative statistical graphs.

It is built on top of Matplotlib, which means it uses Matplotlib to draw graphs but provides a simpler and more attractive interface.

Using Seaborn, we can create:

✔ Line Charts

✔ Bar Charts

✔ Scatter Plots

✔ Box Plots

✔ Violin Plots

✔ Histograms

✔ Heatmaps

✔ Pair Plots

✔ Distribution Plots

Seaborn makes data visualization easier and more visually appealing.

# History of Seaborn

# Import Library

```python
# Import Seaborn

import seaborn as sns

import matplotlib.pyplot as plt
```

# Load Built-in Dataset

```python
# Load sample dataset

tips = sns.load_dataset("tips")

print(tips.head())
```

# Display Dataset Information

```python
tips.info()
```

# Line Plot

```python
sns.lineplot(data=tips,x="total_bill",y="tip")

plt.show()
```

# Scatter Plot

```python
sns.scatterplot(data=tips,x="total_bill",y="tip")

plt.show()
```

# Bar Plot

```python
sns.barplot(data=tips,x="day",y="total_bill")

plt.show()
```

# Count Plot

```python
sns.countplot(data=tips,x="day")

plt.show()
```

# Histogram

```python
sns.histplot(data=tips,x="total_bill")

plt.show()
```

# Box Plot

```python
sns.boxplot(data=tips,x="day",y="total_bill")

plt.show()
```

# Violin Plot

```python
sns.violinplot(data=tips,x="day",y="total_bill")

plt.show()
```

# Strip Plot

```python
sns.stripplot(data=tips,x="day",y="total_bill")

plt.show()
```

# Swarm Plot

```python
sns.swarmplot(data=tips,x="day",y="total_bill")

plt.show()
```

# Pair Plot

```python
sns.pairplot(tips)

plt.show()
```

# Heatmap

```python
corr = tips.corr(numeric_only=True)

sns.heatmap(corr,annot=True,cmap="coolwarm")

plt.show()
```

# KDE Plot

```python
sns.kdeplot(data=tips,x="total_bill",fill=True)

plt.show()
```

# Regression Plot

```python
sns.regplot(data=tips,x="total_bill",y="tip")

plt.show()
```

# Joint Plot

```python
sns.jointplot(data=tips,x="total_bill",y="tip")

plt.show()
```

# Rug Plot

```python
sns.rugplot(data=tips,x="total_bill")

plt.show()
```

# Theme Style

```python
sns.set_style("darkgrid")

sns.barplot(data=tips,x="day",y="total_bill")

plt.show()
```

# Color Palette

```python
sns.set_palette("Set2")

sns.barplot(data=tips,x="day",y="total_bill")

plt.show()
```

# Title and Labels

```python
plt.figure(figsize=(8,5))

sns.scatterplot(data=tips,x="total_bill",y="tip")

plt.title("Restaurant Tips")

plt.xlabel("Total Bill")

plt.ylabel("Tip")

plt.show()
```
