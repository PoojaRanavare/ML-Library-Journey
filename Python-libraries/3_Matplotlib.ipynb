#3. History of Matplotlib

## Before Matplotlib

Before Matplotlib was introduced, Python did not have a powerful and easy-to-use plotting library.

Scientists and engineers often used software like MATLAB to create graphs. Although MATLAB was powerful, it was commercial software that required a paid license.

Many researchers wanted a free alternative that could work directly with Python.

## Who Invented Matplotlib?

Matplotlib was created by **John D. Hunter**, an American neurobiologist and software developer.

He began developing Matplotlib around **2002** while working on neuroscience research. His goal was to create a free plotting library with a style similar to MATLAB.

The first public release of Matplotlib quickly gained popularity because it allowed users to generate high-quality scientific graphs directly from Python.

Today, Matplotlib is maintained by an active open-source community and is one of the most widely used visualization libraries in the Python ecosystem.

# 📊 Chapter 9 : Multiple Line Graphs and Legend

## Introduction

Until now, we have plotted only one line on a graph.

However, in real life, we often need to compare two or more datasets.

### Examples

• Sales of Company A vs Company B

• Temperature of Two Cities

• Student A Marks vs Student B Marks

• Petrol Price vs Diesel Price

• Rainfall of Different States

Instead of creating separate graphs, Matplotlib allows us to display multiple lines on the same graph.

In this chapter, you will learn:

````python
import matplotlib.pyplot as plt

days = [1,2,3,4,5]

companyA = [120,150,170,190,220]

companyB = [100,130,160,210,250]

plt.plot(days, companyA)

plt.plot(days, companyB)

plt.show()
````


## Explanation

The graph contains two different lines.

The first line represents Company A.

The second line represents Company B.

Although both lines are displayed correctly, it is difficult to identify which line belongs to which company.

To solve this problem, we use **labels** and a **legend**.


````python
import matplotlib.pyplot as plt

days=[1,2,3,4,5]

companyA=[120,150,170,190,220]

companyB=[100,130,160,210,250]

plt.plot(days, companyA, label="Company A")

plt.plot(days, companyB, label="Company B")

plt.legend()

plt.show()
````


# Function : label

## Definition

The **label** parameter assigns a name to each line.

This name is displayed inside the legend.

## Syntax

```python
plt.plot(x, y, label="Company A")
```

Without labels, the legend cannot identify each dataset.



````python
plt.legend()
````


# Function : plt.legend()

## Definition

The **legend()** function displays the names of all plotted datasets.

A legend acts like a key or guide that helps readers understand which line represents which dataset.

## Syntax

```python
plt.legend()
```

## Why is it Important?

Imagine a graph containing five different lines.

Without a legend, nobody can identify each line.

Therefore, adding a legend is considered a good practice in professional graphs.



````python
#Example
import matplotlib.pyplot as plt

days=[1,2,3,4,5]

companyA=[120,150,170,190,220]

companyB=[100,130,160,210,250]

plt.plot(
    days,
    companyA,
    color="blue",
    marker="o",
    linewidth=3,
    label="Company A"
)

plt.plot(
    days,
    companyB,
    color="red",
    marker="s",
    linewidth=3,
    label="Company B"
)

plt.title("Weekly Sales Comparison")

plt.xlabel("Days")

plt.ylabel("Sales")

plt.grid(True)

plt.legend()

plt.show()
````


# Output Explanation

Blue Line → Company A

Red Line → Company B

Circle Marker → Company A

Square Marker → Company B

The graph now looks much more professional and is easy to understand.


````python
plt.legend(loc="upper left")
````

## What is Figure Size?

When Matplotlib creates a graph, it automatically uses a default size.

Sometimes the graph appears too small or too large.

To control the width and height of the graph, Matplotlib provides the **figure()** function.

### Definition

`plt.figure()` creates a new figure (drawing area) and allows us to set its size.

### Syntax

```python
plt.figure(figsize=(width, height))
```

The values are measured in **inches**.


````python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [15,25,35,45,55]

plt.figure(figsize=(8,5))

plt.plot(x,y)

plt.title("Figure Size Example")

plt.show()
````


# xticks() and yticks()

• Font Size

• Rotation

````python
import matplotlib.pyplot as plt

days=[1,2,3,4,5]

sales=[120,150,170,200,250]

plt.figure(figsize=(10,6),dpi=120,facecolor="white")

plt.plot(
    days,
    sales,
    color="royalblue",
    linewidth=3,
    marker="o",
    markersize=8,
    label="Sales"
)

plt.title(
    "Weekly Sales Report",
    fontsize=18,
    fontweight="bold",
    color="darkblue"
)

plt.xlabel("Days",fontsize=14)

plt.ylabel("Sales",fontsize=14)

plt.grid(
    linestyle="--",
    alpha=0.5
)

plt.legend()

plt.show()
````

#Basic Histogram

#What is a Pie Chart?

A **Pie Chart** is used to show how different categories contribute to a whole.

Each category is represented by a slice of the circle.

### Real-Life Examples

- Monthly Expenses
- Company Market Share
- Election Results
- Student Grade Distribution

### Function Used

```python

```


#Basic Pie Chart

````python
import matplotlib.pyplot as plt

expenses = [35, 25, 20, 20]
labels = ["Food", "Transport", "Shopping", "Others"]

plt.pie(expenses, labels=labels)

plt.title("Monthly Expenses")

plt.show()
````

#Show Percentage  

````python
plt.pie(
    expenses,
    labels=labels,
    autopct="%1.1f%%"
)

plt.title("Expense Distribution")

plt.show()
````

#Explode a slice

````python
explode = [0.1, 0, 0, 0]

plt.pie(
    expenses,
    labels=labels,
    explode=explode,
    autopct="%1.1f%%"
)

plt.title("Highlighted Food Expense")

plt.show()
````

#Add colors

````python
colors = ["gold", "skyblue", "lightgreen", "pink"]

plt.pie(
    expenses,
    labels=labels,
    colors=colors,
    autopct="%1.1f%%"
)

plt.title("Colorful Pie Chart")

plt.show()
````


# Common Parameters

| Parameter | Description |
|-----------|-------------|
| labels | Names of slices |
| colors | Changes slice colors |
| autopct | Displays percentages |
| explode | Separates a slice |
| shadow | Adds shadow effect |

Example:

#  What are Subplots?

A **subplot** allows you to display **multiple graphs in a single figure**.

Instead of creating separate graph windows, different graphs are arranged together.

### Real-Life Examples

- Sales Analysis Dashboard
- Student Performance Report
- Weather Report
- Stock Market Dashboard

### Function Used

```python
plt.subplot(rows, columns, position)
```


````python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y1 = [10,20,30,40]
y2 = [40,30,20,10]

plt.subplot(1,2,1)
plt.plot(x, y1)
plt.title("Graph 1")

plt.subplot(1,2,2)
plt.plot(x, y2)
plt.title("Graph 2")

plt.show()
````


# Understanding subplot()

```python
plt.subplot(1,2,1)
```

| Value | Meaning |
|--------|---------|
| 1 | Number of rows |
| 2 | Number of columns |
| 1 | Position of the graph |

Example:

subplot(2,2,3)

Means:

2 Rows

2 Columns

3rd Position



````python
plt.figure(figsize=(8,6))

plt.subplot(2,2,1)
plt.plot([1,2,3],[2,4,6])
plt.title("Graph 1")

plt.subplot(2,2,2)
plt.bar(["A","B","C"],[3,5,2])
plt.title("Graph 2")

plt.subplot(2,2,3)
plt.scatter([1,2,3],[5,3,6])
plt.title("Graph 3")

plt.subplot(2,2,4)
plt.pie([40,30,30],labels=["A","B","C"])
plt.title("Graph 4")

plt.tight_layout()

plt.show()
````


# Function : tight_layout()
