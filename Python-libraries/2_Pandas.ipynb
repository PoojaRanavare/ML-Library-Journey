## What is Pandas?

Pandas is an open-source Python library for storing, cleaning, analyzing and manipulating tabular data. Main structures: Series and DataFrame.

## History

First released in 2008 by Wes McKinney to simplify fast data analysis in Python.

## Why Pandas?

It makes reading CSV/Excel files, cleaning data, filtering, grouping and analysis easy.

## Features

- Fast
- Data cleaning
- File support
- Grouping
- Statistics
- Works with NumPy

Import pandas and check version.

```python
import pandas as pd
print(pd.__version__)
```

Create a Series.

```python
s=pd.Series([10,20,30,40])
print(s)
```

Create a DataFrame.

```python
data={'Name':['Rahul','Priya','Amit'],'Age':[20,19,21],'Marks':[88,91,75]}
df=pd.DataFrame(data)
df
```

Display first rows.

```python
df.head()
```

Display last rows.

```python
df.tail()
```

Dataset information.

```python
df.info()
```

Statistical summary.

```python
df.describe()
```

Add a new column.

```python
df['Grade']=['A','A+','B']
df
```

Filter rows.

```python
df[df['Marks']>80]
```

Sort data.

```python
df.sort_values('Marks')
```

Count missing values.

```python
df.isnull().sum()
```

Mean marks.

```python
df['Marks'].mean()
```

Delete Column

```python
df.drop("Age",axis=1,inplace=True)

print(df)
```

Rename Column

```python
df.rename(columns={"Marks":"Score"},inplace=True)

print(df)
```

Maximum

```python
df["Score"].max()
```

Minimum

```python
df["Score"].min()
```

Merge DataFrames

```python
df1 = pd.DataFrame({
    "ID":[1,2],
    "Name":["Rahul","Priya"]
})

df2 = pd.DataFrame({
    "ID":[1,2],
    "Marks":[90,95]
})

result = pd.merge(df1,df2,on="ID")

print(result)
```
